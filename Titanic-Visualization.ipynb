```python
def male_female_child(passeneger):
    age,sex = passeneger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child, axis=1)
```

```python
#who was alone and who was with a family
```

```python
titanic_df['SibSp'].value_counts()
```

```python
def family_or_alone(passenger):
    SibSp, Parch = passenger
    if SibSp != 0 or Parch !=0:
        return 'With Family'
    else:
        return 'Alone'
```

```python
titanic_df['Alone']=titanic_df[['SibSp','Parch']].apply(family_or_alone, axis=1)
```

```python
titanic_df.head()
```

```python
sns.catplot(data=titanic_df, kind='count', x='Alone')
plt.show()
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no',1:'yes'})
```
